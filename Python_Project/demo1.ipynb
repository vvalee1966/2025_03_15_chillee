- 建立`正弦波`圖表範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 軸數據，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2*np.pi, 0.1)

# 計算 y 軸數據，使用 sin() 函數
y = np.sin(x)

# 使用 matplotlib 繪製圖表
plt.plot(x, y)  # 繪製 x 和 y 的關係圖，形成正弦波圖表

# 設定圖表標題和軸標籤
plt.title('sin plot')  # 設定圖表標題
plt.xlabel('x (弧度)')  # 設定 x 軸標籤
plt.ylabel('sin(x)')  # 設定 y 軸標籤

# 顯示圖表
plt.show()  # 顯示繪製好的圖表
```
